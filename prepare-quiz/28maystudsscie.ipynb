```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Bikin data koordinat X (misal dari angka -10 sampai 10 sebanyak 50 titik)
x = np.linspace(-10, 10, 50)

# 2. Tentukan rumus garis linear (Misal: y = 2x + 3)
# m = 2, c = 3
y = 2 * x + 3

# 3. Plot grafik garisnya
plt.figure(figsize=(6, 6)) # Bikin kanvas kotak (6x6 inci)
plt.plot(x, y, label='Garis y = 2x + 3', color='red', linewidth=2)

# 4. Tambahin garis bantu sumbu X dan Y (biar kelihatan titik (0,0) di tengah)
plt.axhline(0, color='black', linewidth=1, linestyle='--') # Garis horizontal horizontal
plt.axvline(0, color='black', linewidth=1, linestyle='--') # Garis vertikal

# 5. Aksesoris grafik
plt.title('Grafik Garis Linear (Persamaan Linear)', fontsize=12)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.grid(True, alpha=0.3) # Kasih kotak-kotak samar di background
plt.legend()

# 6. Tampilkan
plt.show()
```
